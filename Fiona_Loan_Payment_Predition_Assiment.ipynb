# **Cleaning of data**

```python
#let's drop column with Nan value
clean_loans = loan.dropna()
```

```python
percent_missing = clean_loans.isnull().sum() * 100 / len(clean_loans)
missing_value_df = pd.DataFrame({'column_name': clean_loans.columns,
                                 'percent_missing': percent_missing})

missing_value_df
```

```python
clean_loans['loan_status'].value_counts()
```

```python
clean_loans['loan_status'] = clean_loans['loan_status'].str.replace('Does not meet the credit policy. Status:Fully Paid', 'Fully Paid')
clean_loans['loan_status'] = clean_loans['loan_status'].str.replace('Does not meet the credit policy. Status:Charged Off', 'Charged Off')
# clean_loans['loan_status'] = clean_loans['loan_status'].str.replace('Does not meet the credit policy. Status:Charged Off', 'Charged Off ')
```

```python
clean_loans['term'] = clean_loans['term'].str.replace('months', '')
clean_loans['int_rate'] = clean_loans['int_rate'].str.replace('%', '')
```

```python
data = clean_loans[['loan_amnt','funded_amnt_inv', 'term','int_rate','installment','grade', 'sub_grade','home_ownership', 'annual_inc','loan_status','purpose','inq_last_6mths','open_acc','total_pymnt','total_pymnt_inv','total_rec_prncp','total_rec_int']]
```

```python
data
```

```python
df_dummies = pd.get_dummies(data, columns=['grade','sub_grade','home_ownership','purpose'])
df_dummies
```

```python
# let us drop column that has loaan status of current
current = df_dummies.loc[df_dummies['loan_status'] == "Current"]
current.index
```

```python
default = df_dummies.loc[df_dummies['loan_status'] == "Default"]
default.index
```

```python
new_data = df_dummies.drop(current.index)
new_data = new_data.drop(default.index)
new_data
```

```python
new_data['loan_status'].value_counts()
```

```python
#Group variables
new_data['loan_status'] = new_data['loan_status'].map({'Fully Paid': 0, 'Charged Off': 1, 'Late (31-120 days)': 1, 'In Grace Period': 1, 'Late (16-30 days)': 1,})
new_data 
```

# **Build a machine learning  model to predict our result**

```python
X = new_data.drop(['loan_status'],axis=1).values
y = new_data['loan_status'].values
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, shuffle=True, random_state=42)
print('The number of training samples: {}\nThe number of testing samples: {}'.format(X_train.shape[0], X_test.shape[0]))
```

```python
# Evaluating our model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,mean_squared_error
lrr = LogisticRegression(solver='liblinear')
lrr.fit(X_train, y_train)
y3_pred = lrr.predict(X_test)
mean_squared_error(y_test,y_test)
```

```python
accuracy_score(y_test,y3_pred)
```
